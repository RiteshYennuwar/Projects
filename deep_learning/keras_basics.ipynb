```python
# relu - rectified linear unit

# if x>0 , x
# if x<0 , 0
# x = [1,2,3,4,-1,-4,-9]
# y = [1,2,3,4,0,0,0]

#wap to implement relu
```

```python
x = np.linspace(-10,10,100)

y = []

for i in x:
    if i>=0:
        y.append(i)
    else:
        y.append(0)

plt.plot(x,y)
plt.show()
```

```python
import tensorflow as tf
```

```python
(x_train,y_train),(x_test,y_test) = tf.keras.datasets.boston_housing.load_data(
    path="boston_housing.npz", test_split=0.2, seed=113)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#dataframe for traning data
import pandas as pd
df = pd.DataFrame(x_train)
df['Target'] = y_train
df.head()
```

```python
# step 2 : 
from sklearn.preprocessing import StandardScaler
s = StandardScaler()
x_train_scaled = s.fit_transform(x_train)
x_test_scaled = s.fit_transform(x_test)
```

```python
# step 3 : NN Architeture
# no of nerouns
# input layer : no of input values = 13
# hidden layer : in between the values of ip and op = 7
# output layer : 1 mostly for regression = 1
```

```python
#NN_SVG
```

```python
# parameters : weight and bias
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(7,input_shape=(13,), activation='relu'))
model.add(tf.keras.layers.Dense(1))
```

```python
model.summary()
```

```python
# output layer
# wieghts (7*1)
# bias(1)
# parms = (7*1)+1
(7*1)+1
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model.fit(x_train,y_train,epochs=500)
```

```python
y_pred = model.predict(x_test)
y_pred[0:2]
```
